```python
#Newton-Cotes Quadrature Formula
import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate
def f(x):
    return np.sin(x)

# Trapezoidal rule implementation
def trapezoidal_rule(f, a, b):
    return (b - a) * (f(a) + f(b)) / 2

# Simpson's rule implementation
def simpsons_rule(f, a, b):
    c = (a + b) / 2
    return (b - a) * (f(a) + 4*f(c) + f(b)) / 6

# Higher-order Newton-Cotes formula (n=3, Simpson's 3/8 rule)
def simpsons_3_8_rule(f, a, b):
    h = (b - a) / 3
    return (3 * h / 8) * (f(a) + 3*f(a + h) + 3*f(a + 2*h) + f(b))

# Define integration limits
a = 0
b = np.pi

# Compute the integral using different methods
integral_trapezoidal = trapezoidal_rule(f, a, b)
integral_simpsons = simpsons_rule(f, a, b)
integral_simpsons_3_8 = simpsons_3_8_rule(f, a, b)

print(f"Trapezoidal rule: {integral_trapezoidal}")
print(f"Simpson's rule: {integral_simpsons}")
print(f"Simpson's 3/8 rule: {integral_simpsons_3_8}")

# Plot the function and integration points
x = np.linspace(a, b, 100)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = sin(x)')
plt.fill_between(x, y, alpha=0.2)
plt.scatter([a, b], [f(a), f(b)], color='red', label='Trapezoidal points')
plt.scatter([a, (a + b) / 2, b], [f(a), f((a + b) / 2), f(b)], color='green', label="Simpson's points")
plt.scatter([a, a + (b - a) / 3, a + 2 * (b - a) / 3, b], [f(a), f(a + (b - a) / 3), f(a + 2 * (b - a) / 3), f(b)], color='blue', label="Simpson's 3/8 points")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Newton-Cotes Quadrature')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Bool;s rule
import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate
def f(x):
    return np.sin(x)

# Boole's rule implementation
def booles_rule(f, a, b):
    h = (b - a) / 4
    return (2 * h / 45) * (7*f(a) + 32*f(a + h) + 12*f(a + 2*h) + 32*f(a + 3*h) + 7*f(b))

# Define integration limits
a = 0
b = np.pi

# Compute the integral using Boole's rule
integral_booles = booles_rule(f, a, b)

print(f"Boole's rule: {integral_booles}")

# Plot the function and integration points
x = np.linspace(a, b, 100)
y = f(x)
x_points = np.array([a, a + (b - a) / 4, a + 2 * (b - a) / 4, a + 3 * (b - a) / 4, b])
y_points = f(x_points)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = sin(x)')
plt.fill_between(x, y, alpha=0.2)
plt.scatter(x_points, y_points, color='red', label="Boole's points")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Boole's Rule Quadrature")
plt.legend()
plt.grid(True)
plt.show()
```

```python
###Romberg

import numpy as np
import matplotlib.pyplot as plt

# Define the function to integrate
def f(x):
    return np.sin(x)

# Trapezoidal rule implementation
def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    result = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        result += f(a + i * h)
    return result * h

# Romberg integration implementation
def romberg_integration(f, a, b, max_k):
    R = np.zeros((max_k, max_k), dtype=float)
    for k in range(max_k):
        R[k, 0] = trapezoidal_rule(f, a, b, 2**k)
        for j in range(1, k+1):
            R[k, j] = (4**j * R[k, j-1] - R[k-1, j-1]) / (4**j - 1)
    return R

# Define integration limits and accuracy
a = 0
b = np.pi
max_k = 5  # Number of rows/columns in the Romberg table

# Compute the Romberg integration table
R = romberg_integration(f, a, b, max_k)

# Print the Romberg table
print("Romberg integration table:")
print(R)

# Plot the convergence of the Romberg integration
plt.figure(figsize=(10, 6))
for i in range(max_k):
    plt.plot(range(i+1), R[i, :i+1], 'o-', label=f'k={i}')

plt.xlabel('j')
plt.ylabel('R(k, j)')
plt.title('Convergence of Romberg Integration')
plt.legend()
plt.grid(True)
plt.show()

# The best estimate of the integral is in the bottom-right corner of the Romberg table
best_estimate = R[max_k-1, max_k-1]
print(f"Best estimate of the integral: {best_estimate}")
```

```python
#Euler maclauren
import numpy as np
from scipy.special import bernoulli

# Define the function to integrate
def f(x):
    return np.sin(x)

# Euler-Maclaurin formula implementation
def euler_maclaurin(f, a, b, m=2):
    # Compute the integral part
    integral_approx = np.trapz([f(x) for x in np.linspace(a, b, 1000)], dx=(b-a)/1000)
    
    # Compute the correction terms
    correction_terms = (f(a) + f(b)) / 2
    for k in range(1, m+1):
        B2k = bernoulli(2*k)
        f_deriv_b = np.diff([f(b + i * 1e-5) for i in range(2*k + 1)], 2*k)[0]
        f_deriv_a = np.diff([f(a + i * 1e-5) for i in range(2*k + 1)], 2*k)[0]
        correction_terms += B2k / (2 * k) * (f_deriv_b - f_deriv_a) / (1e-5 ** (2*k))

    return integral_approx + correction_terms

# Define integration limits
a = 0
b = np.pi

# Compute the integral using the Euler-Maclaurin formula
integral_em = euler_maclaurin(f, a, b)

print(f"Euler-Maclaurin formula: {integral_em}")

# Comparison with the exact integral
exact_integral = 2  # since integral of sin(x) from 0 to pi is 2
print(f"Exact integral: {exact_integral}")
print(f"Error: {abs(integral_em - exact_integral)}")
```

```python
import numpy as np
from scipy.special import bernoulli
from scipy.misc import derivative

# Define the function to integrate
def f(x):
    return np.sin(x)

# Euler-Maclaurin formula implementation
def euler_maclaurin(f, a, b, m=2):
    # Compute the integral part using trapezoidal rule
    integral_approx = np.trapz([f(x) for x in np.linspace(a, b, 1000)], dx=(b-a)/1000)
    
    # Compute the correction terms
    correction_terms = (f(a) + f(b)) / 2
    for k in range(1, m+1):
        B2k = bernoulli(2*k)
        # Ensure that order is at least 2k + 1
        f_deriv_b = derivative(f, b, dx=1e-5, n=2*k-1, order=2*k+1)
        f_deriv_a = derivative(f, a, dx=1e-5, n=2*k-1, order=2*k+1)
        correction_terms += (B2k / (2*k)) * (f_deriv_b - f_deriv_a)

    return integral_approx + correction_terms

# Define integration limits
a = 0
b = np.pi

# Compute the integral using the Euler-Maclaurin formula
integral_em = euler_maclaurin(f, a, b)

print(f"Euler-Maclaurin formula: {integral_em}")

# Comparison with the exact integral
exact_integral = 2  # since integral of sin(x) from 0 to pi is 2
print(f"Exact integral: {exact_integral}")
print(f"Error: {abs(integral_em - exact_integral)}")
```

```python
#Gaussian Integration
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.legendre import leggauss

# Define the function to integrate
def f(x):
    return np.sin(x)

# Gauss-Legendre Quadrature implementation
def gauss_legendre_quadrature(f, a, b, n):
    # Get the roots and weights for the n-point Gauss-Legendre quadrature
    [x, w] = leggauss(n)
    
    # Change of interval from [-1, 1] to [a, b]
    t = 0.5 * (x + 1) * (b - a) + a
    integral = 0.5 * (b - a) * np.sum(w * f(t))
    
    return integral

# Define integration limits
a = 0
b = np.pi

# Number of points for Gauss-Legendre Quadrature
n = 5

# Compute the integral using Gauss-Legendre Quadrature
integral_gauss_legendre = gauss_legendre_quadrature(f, a, b, n)

print(f"Gauss-Legendre Quadrature (n={n}): {integral_gauss_legendre}")

# Comparison with the exact integral
exact_integral = 2  # since integral of sin(x) from 0 to pi is 2
print(f"Exact integral: {exact_integral}")
print(f"Error: {abs(integral_gauss_legendre - exact_integral)}")

# Visualize the quadrature points and weights
[x, w] = leggauss(n)
t = 0.5 * (x + 1) * (b - a) + a

# Plot the function and quadrature points
x_plot = np.linspace(a, b, 1000)
y_plot = f(x_plot)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='f(x) = sin(x)')
plt.scatter(t, f(t), color='red', zorder=5, label='Quadrature points')
plt.vlines(t, ymin=0, ymax=f(t), color='red', linestyle='dashed', zorder=5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f"Gauss-Legendre Quadrature (n={n})")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import sympy as sp

# Define the symbols
x = sp.symbols('x')
C1, C2 = sp.symbols('C1 C2')

# Define the function and the ODE
y = sp.Function('y')(x)
ode = sp.Eq(y.diff(x, x) - 3*y.diff(x) + 2*y, sp.exp(2*x))

# Solve the ODE
solution = sp.dsolve(ode)
print("General solution:")
sp.pprint(solution)

# Verify the particular solution
particular_solution = sp.exp(2*x) * x / 3
ode_lhs = y.diff(x, x) - 3*y.diff(x) + 2*y
ode_check = ode_lhs.subs(y, particular_solution).simplify()
print("\nParticular solution verification:")
sp.pprint(ode_check)
```

```python
import numpy as np
from scipy.integrate import dblquad

# Define the function to integrate
def integrand(x, y):
    return np.sin(x) * np.cos(y)

# Define the limits of integration
x_lower = 0
x_upper = np.pi
y_lower = 0
y_upper = np.pi / 2

# Perform the double integration using scipy's dblquad
result, error = dblquad(integrand, y_lower, y_upper, lambda y: x_lower, lambda y: x_upper)

print(f"Numerical double integral: {result}")
print(f"Estimated error: {error}")
```
